```python
self_harm_all_df = pd.DataFrame(columns=['country', 'self_harm'])


self_harm_all_df = self_harm_all_df.append(df.loc[ (df['year'] == 1999) & (~df['country'].isin(unions)) , ['country', 'self_harm']])

self_harm_all_df.reset_index(inplace = True, drop=True)

self_harm_all_df['self_harm'] = self_harm_all_df['self_harm'] / (dfp['population'] / 100)  

self_harm_all_df = self_harm_all_df.loc[(self_harm_all_df['self_harm'] > 0.02),  ['country', 'self_harm']]

pd.set_option('display.float_format', '{:.4f}'.format)


self_harm_all_df.sort_values(by='country', ascending=True)
```

```python
plt.figure(figsize=(20,90))
sns.barplot(x="self_harm", y="country",data=self_harm_all_df)
plt.title('Количество самоубийств за 1999 год в мире') 
plt.ylabel('Страна')
plt.xlabel('Процент самоубийств по отношению к населению за 1990г.')
```

```python
self_harm_chine_df = pd.DataFrame(columns=['country', 'self_harm'])
self_harm_chine_df = self_harm_chine_df.append(df.loc[ (df['year'] == 1999) & (df['country'].isin(['China', 'India'])) , ['country', 'self_harm']])
  
self_harm_chine_df.reset_index(inplace = True, drop=True)
self_harm_chine_df

self_harm_chine_df.loc[self_harm_chine_df['country'] == 'China', 'self_harm'] = 198900 / (1253000000 / 100)
self_harm_chine_df.loc[self_harm_chine_df['country'] == 'India', 'self_harm'] = 195957 / (1038000000 / 100)
# = self_harm_all_df['self_harm'] / (dfp['population'] / 100)  

plt.figure(figsize=(20,4))
sns.barplot(x="self_harm", y="country",data=self_harm_chine_df)
plt.title('Количество самоубийств за 1999 год в Китае и Индии') 
plt.ylabel('Страна')
plt.xlabel('Количество самоубийств')
```
